```python
df["Kat No"].unique()
```

```python
df.loc[df["Kat No"]=="Dubleks"]
```

```python
df.drop(df[df["Kat No"]=="Dubleks"].index, inplace = True)
```

```python
df.loc[df["Kat No"]=="Tripleks"]
```

```python
df.drop(df[df["Kat No"]=="Tripleks"].index, inplace = True)
```

```python
df["Kat No"].unique()
```

```python
df["Kat No"]=df["Kat No"].replace(["YüksekGiriş","None","GirişKatı","BahçeKatı","Zemin","BodrumveZemin","VillaKatı","YarıBodrum",'-1ve0'],"0",regex=True)
df["Kat No"]=df["Kat No"].str.replace("Katı","")
df["Kat No"]=df["Kat No"].str.replace(". Kat","")
df["Kat No"]=df["Kat No"].str.replace(".Kat","")
df["Kat No"]=df["Kat No"].str.replace("Kot 3","-3")
df["Kat No"]=df["Kat No"].str.replace("Kot 2","-2")
df["Kat No"]=df["Kat No"].str.replace("Kot 1","-1")
df["Kat No"]=df["Kat No"].str.replace("Kot3","-3")
df["Kat No"]=df["Kat No"].str.replace("Kot2","-2")
df["Kat No"]=df["Kat No"].str.replace("Kot1","-1")
df["Kat No"]=df["Kat No"].replace(["EnÜstKat","EnÜs"],"4")
df["Kat No"]=df["Kat No"].str.replace("Çatı","5")
df["Kat No"]=df["Kat No"].replace(["TerasKatı","Teras"],"4")
df["Kat No"]=df["Kat No"].replace(["Ar","AraKat"],"2")
df["Kat No"]=df["Kat No"].str.replace("Bodrum","-1")
df["Kat No"]=df["Kat No"].fillna("0")
df["Kat No"]=df["Kat No"].astype(int)
```
